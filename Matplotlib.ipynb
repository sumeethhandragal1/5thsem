```python
x = np.arange(1,16)    
y = x**2    
z = (y/2)+40    
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)   
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)   
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)   
ax1.legend()    
ax2.legend()
```

```python
x = np.arange(1,16)    
y = x**2    
z = (y/2)+40    
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_axes([0,0,0.5,0.5])
ax2 = fig.add_axes([0.6,0,0.5,0.5])
ax3 = fig.add_axes([0,0.6,0.5,0.5])
ax4 = fig.add_axes([0.6,0.6,0.5,0.5])

ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)   
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15) 

ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)  

ax3.plot(x,y,color="b", label='line3' )
ax3.set_title('Title for Plot 3',fontsize=15)   
ax3.set_xlabel('Plot 3 - X axis',fontsize=15)    
ax3.set_ylabel('Plot 3 - Y axis',fontsize=15) 
 
ax4.plot(x,y,color="y", label='line4' )
ax4.set_title('Title for Plot 4',fontsize=15)   
ax4.set_xlabel('Plot 4 - X axis',fontsize=15)    
ax4.set_ylabel('Plot 4 - Y axis',fontsize=15) 

ax1.legend()    
ax2.legend()
ax3.legend() 
ax4.legend() 
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('dataset/auto_mpg.csv')
df.info()  
```

```python
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)
df.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= df['horsepower']
y= df['mpg']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y)
```

```python
df.groupby('model_year').count()['name']
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('model_year').count()[['name']]
df_new.reset_index(inplace = True)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df_new['model_year'],df_new['name'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.hist(df['horsepower'], bins=20)
```

```python
new_df = df.groupby('origin').count()[['name']]
new_df.reset_index(inplace=True)
new_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
new_df = df.groupby('origin').count()[['name']]
new_df.reset_index(inplace=True)
colors = ['red', 'green', 'blue' ]
explode = [0.05, 0, 0]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.pie(new_df['name'], labels=new_df['origin'], colors = colors, explode = explode, autopct='%0.1f%%')
```
